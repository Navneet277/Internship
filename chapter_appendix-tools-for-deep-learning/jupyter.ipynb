## Problem Statement
Mental stress is a growing concern among students, affecting their academic performance, physical health, and overall well-being. Stress can stem from academic pressure, financial issues, peer relationships, and personal factors. This study aims to analyze the impact of mental stress on students' academic performance and examine the effectiveness of various coping mechanisms using exploratory data analysis (EDA).

### Objectives:
- Identify key factors contributing to mental stress among students.
- Understand the relationship between stress levels and academic performance.
- Examine the role of coping mechanisms in stress management.
- Provide insights to help educational institutions and mental health professionals support students better.

## Description of the Dataset
This dataset contains **760 student records** with **20 features**, categorized as follows:

- **Demographic Information**:
  - Age
  - Gender
  - Education level

- **Academic Performance**:
  - GPA (Grade Point Average)
  - Study Hours per week
  - Exam Scores

- **Stress Indicators**:
  - Mental Stress Level (on a scale)
  - Sleep Duration (in hours)
  - Anxiety and Depression levels

- **Coping Mechanisms**:
  - Physical Exercise (frequency)
  - Social Support (family & friends)
  - Participation in Counseling Sessions
  - Mindfulness and Meditation Practices

### Data Source:
The dataset was obtained from a student mental health survey conducted in multiple educational institutions.

## Methodology Used
To explore the dataset effectively, we follow a structured **5-step EDA approach**:

### 1. Data Collection
- The dataset is obtained from surveys and academic records of students.

### 2. Data Cleaning & Preprocessing
- Handling **missing values** (imputation or removal)
- Checking for **duplicates** and dropping them
- Converting **data types** to appropriate formats
- Removing outliers using statistical methods (IQR, Z-score)

```python
# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
df = pd.read_csv("/content/Student_Mental_Stress_and_Coping_Mechanisms.csv")

# Display first few rows
display(df.head())

# Check for missing values
missing_values = df.isnull().sum()
print("Missing values per column:")
print(missing_values)

# Remove duplicate rows
df = df.drop_duplicates()

# Convert categorical variables into appropriate data types
if 'Education Level' in df.columns:
    df['Education Level'] = df['Education Level'].astype('category')
else:
    print("Warning: 'Education Level' column not found in the DataFrame.")
    # Inspect the DataFrame columns to identify the correct name
    print("Available columns:", df.columns)

### 3. Exploratory Data Analysis (EDA)
# Summary statistics**: Mean, median, standard deviation, etc.
# Distribution Analysis**: Histograms, KDE plots
# Correlation Analysis**: Heatmaps
# Outlier Detection**: Boxplots
# Comparative Analysis**: Pair plots and grouped boxplots

# Display summary statistics
display(df.describe())

#### 3.1 Distribution of Stress Levels

plt.figure(figsize=(8, 4))
sns.histplot(df['Mental Stress Level'], kde=True, bins=20, color='red')
plt.title("Distribution of Mental Stress Levels")
plt.xlabel("Stress Level")
plt.ylabel("Frequency")
plt.show()

#### 3.2 Correlation Between Features

plt.figure(figsize=(10, 6))
# Exclude 'Student ID' column from correlation calculation
numerical_df = df.select_dtypes(include=np.number)  # Select only numerical columns
sns.heatmap(numerical_df.corr(), annot=True, cmap="coolwarm", linewidths=0.5)
plt.title("Correlation Heatmap of Features")
plt.show()


#### 3.3 Stress vs. Academic Performance

plt.figure(figsize=(8, 5))
sns.boxplot(x=df['Mental Stress Level'], y=df['Academic Performance (GPA)'], palette="coolwarm")
plt.title("Mental Stress Level vs. Academic Performance")
plt.xlabel("Mental Stress Level")
plt.ylabel("GPA")
plt.show()


#### 3.4 Impact of Sleep on Stress

plt.figure(figsize=(8, 4))
sns.scatterplot(x=df['Sleep Duration (Hours per night)'], y=df['Mental Stress Level'], hue=df['Gender'], palette="coolwarm")
plt.title("Relationship Between Sleep Duration and Stress Level")
plt.xlabel("Sleep Duration (Hours)")
plt.ylabel("Mental Stress Level")
plt.show()


```

### 4. Observations & Insights
From the above analyses, we derive the following insights:
- **High stress levels are associated with lower academic performance** (negative correlation between stress and GPA).
- **Lack of sleep is a significant contributor to stress**, as seen in the scatterplot.
- **Students who engage in physical exercise and social activities tend to have lower stress levels**.
- **Counseling attendance shows mixed results**, indicating that while therapy helps, additional support systems are required.

### 5. Experimental Results (Using Visualization Tools)
The following visualizations help interpret the dataset:
- **Histogram**: Distribution of stress levels among students.
- **Heatmap**: Correlation between stress, academic performance, and coping strategies.
- **Boxplot**: Stress levels and GPA distribution.
- **Scatter Plot**: Relationship between sleep duration and stress levels.

## Conclusion
- Mental stress negatively impacts academic performance, with high stress levels correlating with lower GPA scores.
- Proper sleep, social support, and physical exercise are effective coping strategies that reduce stress.
- Institutions should encourage stress management programs, including meditation and professional counseling.

### Recommendations
- Introduce structured mental health programs in universities.
- Encourage students to maintain a **balanced sleep schedule**.
- Promote **physical activities** and **social interactions** to alleviate stress.
- Offer **professional counseling services** and periodic stress assessments.

These insights can help educators, psychologists, and policymakers develop targeted interventions for student well-being.
